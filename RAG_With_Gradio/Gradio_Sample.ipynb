Gradio Interface :

The interface class has three core arugments:
    1. Fn: the function to wrap a UI around
    2. Inputs : The Gradio components to use for the input.
    3. Outputs : The Gradio components to use for the output.

```python
def greet(name):
    return "Hellow "+ name + "!"

demo = gr.Interface(fn=greet, inputs="text", outputs="text")
```

```python
demo.launch()
```

```python
def greet(name, intensity):
    return "Hello " *intensity + name + "!"

demo = gr.Interface(
    fn=greet,
    inputs=["text","slider"],
    outputs=["text"]
    )

demo.launch()
```

```python
def random_response(message,history):
    return random.choice(["Yes","No"])
demo = gr.ChatInterface(random_response)
demo.launch()


```

```python
import gradio as gr

def yes_man(message,history):
    if message.endswith("?"):
        return "Yes"
    else:
        return "Ask me something!"

gr.ChatInterface(
    yes_man,
    chatbot=gr.Chatbot(height=300),
    textbox=gr.Textbox(placeholder="Ask me as yes or no question", container=False, scale=7),
    title="Yes Man",
    description="Ask Yes Man any question",
    #theme=gr.themes.Default(),
    examples=["Hellow", "Am I cool?", "Are tomatoes Vegetables?"],
    cache_examples=True,
    #retry_btn=None,
    #undo_btn=None,
    #clear_btn="Clear"
).launch()
```
